## K-Means

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

image_paths = [
    r"C:\Users\fionu\OneDrive\Dokumente\Daten Fiona\AIUB\Code\AIUB-Backup\clustering\Images\k_means_tests_running_bins_ecc\kmeans_2002-2005_k7.png",
    r"C:\Users\fionu\OneDrive\Dokumente\Daten Fiona\AIUB\Code\AIUB-Backup\clustering\Images\k_means_tests_running_bins_a\kmeans_2002-2005_k7.png",
    r"C:\Users\fionu\OneDrive\Dokumente\Daten Fiona\AIUB\Code\AIUB-Backup\clustering\Images\k_means_tests_running_bins_mag\kmeans_2003-2006_k7.png"
]

titles = [
    "Inclination, RAAN, Eccentricity",
    "Inclination, RAAN, Semi Major Axis",
    "Inclination, RAAN, Magnitude"
]

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

for i, (path, title) in enumerate(zip(image_paths, titles)):
    img = mpimg.imread(path)
    axs[i].imshow(img)
    axs[i].axis('off')
    axs[i].set_title(title, fontsize=10)

plt.tight_layout()
plt.show()
```
