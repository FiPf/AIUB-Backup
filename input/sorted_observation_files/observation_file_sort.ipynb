```python
import numpy as np
import os
#COMMENTS NOT CORRECT!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

#purpose of this script: sort the given yearly automation files for GEO, GTO and followup objects. Exclude all other objects. 
#Create a new *.aut file for each category and store it in a folder for each year
#put this script in the same directory as all the aut files, run it

cwd = os.getcwd()  # Get the current working directory (cwd)
files = os.listdir(cwd)  # Get all the files in that directory

# look into current directory and find all out files
# Assumption: The aut files have "Delivery" or "MASTERaut" in the filename
obs_file_list = []
years = []
for file in files:
    if "ogs" in file and "ucorr" in file:
        obs_file_list.append(file)
        with open(file, 'r') as f:
            third_line = f.readlines()[2]
            year = file[3:7]
            years.append(year)

def get_lines(filename: str):
    """Function to extract all lines from a given file

    Args:
        filename (str): name of the file

    Returns:
        lines (np.array): contains the lines as single elements
    """
    with open(filename, "r") as inp:
        lines = inp.readlines()
        lines = np.array(lines)
    return lines

def create_files(year: str, lines: np.array):
    """Creates three files (GEO, GTO, followup), each containing the corresponding elements from lines

    Args:
        year (str): year of the aut file
        lines (np.array): file containing all lines to be sorted
    """
    # Makes a new folder to store the files for one year
    year_dir = os.path.join(cwd, year)
    os.makedirs(year_dir, exist_ok=True)

    gto_path = os.path.join(year_dir, f'GTO_obs_{year}')
    geo_path = os.path.join(year_dir, f'GEO_obs_{year}')
    followup_path = os.path.join(year_dir, f'followup_obs_{year}')
    
    # Open the files in write mode to clear them before rerunning the code
    # otherwise, the files just get extended with every run of the code
    open(gto_path, 'w').close()
    open(geo_path, 'w').close()
    open(followup_path, 'w').close()

    # Copy the title from the aut file into the new files so proof does not get confused
    with open(gto_path, 'a') as GTO, open(geo_path, 'a') as GEO, open(followup_path, 'a') as followup:
        title_lines = lines[:2]
        for line in title_lines:
            followup.write(line)
            GTO.write(line)
            GEO.write(line)
        
        # Sorting    
        def should_write_to_GEO(line):
            keywords = ["meo", "COVI", "MSG", "fup", "Hylas", "covi", "Ekspress", "BRIZM", "trk"] #ignore all objects that don't fall in our categories
            return all(keyword not in line for keyword in keywords)

        for line in lines[2:]:
            if "fol" in line:
                followup.write(line)
            elif "gto" in line:
                GTO.write(line)
            elif should_write_to_GEO(line):
                GEO.write(line)
            #else:
                #GEO.write(line)

# Do the process for all aut files in the directory
for aut_file, year in zip(obs_file_list, years):
    lines = get_lines(aut_file)
    create_files(year, lines)
```
